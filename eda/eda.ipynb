## Transformaciones sobre la tabla cards

### Columnas de tipo texto

```python
columns_to_string = [
    'id',
    'client_id',
    'card_number',
    'cvv'
]
for col in columns_to_string:
    cards[col] = cards[col].astype(str)

cards.dtypes
```

### Columnas de tipo bool

```python
cards.has_chip.value_counts()
```

```python
cards.card_on_dark_web.value_counts()
```

```python

columns_to_bool = [
    'has_chip',
    'card_on_dark_web',
]

map_bool = {'no': False, 'yes': True}
for col in columns_to_bool:
    cards[col] = cards[col].apply(lambda x: x.lower()).map(map_bool)

cards.dtypes
```

### Columnas de tipo numérico

```python
cards.credit_limit.value_counts()
```

```python
columns_to_int = [
    'credit_limit',
]
for col in columns_to_int:
    cards[col] = cards[col].apply(lambda x: int(x.replace('$', '')))

cards.dtypes
```

### Columnas de tipo date

```python
columns_to_date = [
    'expires',
    'acct_open_date',
]
for col in columns_to_date:
    cards[col] = pd.to_datetime(cards[col], format='%m/%Y')

cards.dtypes
```
